## Train

```python
dimensione_input = 640
```

```python
# build from YAML and transfer weights
model = YOLO('yolov8s.pt')

# versione Yolov8small
```

```python
model.train(
    data = "./data_ita.yaml",
    pretrained=True,
    epochs=100,
    imgsz=dimensione_input, 
    #optimizer='SGD',
    optimizer='Adam',
    save=True,
    )
```

```python
metrics = model.val()  # evaluate model performance on the validation set
```
