## Train

```python
dimensione_input = 1280
```

```python
# build from YAML and transfer weights
model = YOLO('yolov8m.pt')

# versione Yolov8medium
```

```python
model.train(
    data = "./data_ita.yaml",
    pretrained=True,
    epochs=200,
    imgsz=dimensione_input, 
    optimizer='SGD',
    lr0 = 1e-3,
    #lrf = 1e-5,
    #optimizer='Adam',
    save=True,
    batch=-1
    )
```

```python
metrics = model.val(batch=16)  # evaluate model performance on the validation set
```
